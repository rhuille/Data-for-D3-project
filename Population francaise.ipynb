```python
from pandas import *
```

```python
data = read_excel("HIST_POP_COM_RP13.xls", encoding="iso-8859-9", header=5 )[[
        "DEPCOM","DEP", "REG", "PMUN13","PMUN08", "PSDC99", "PSDC90", "PSDC82", "PSDC75", "PSDC68", "PSDC62"]]
data = data.merge(read_csv("http://www.insee.fr/fr/methodes/nomenclatures/cog/telechargement/2013/txt/depts2013.txt", sep="\t")[["NCC", "DEP"]])
data.columns = Series(["Communes","Departements", "Regions", "2013", "2008", "1999", "1990", "1982", "1975", "1968", "1962", "Departements_names"])
name = read_excel("table-appartenance-geo-communes-13.xls", header=5)[["CODGEO","LIBGEO"]]
name.columns = Series(["Communes", "Communes_ames"])
data = data.merge(name)
data.head()
```

## By departements

```python
departements = []
for dep in set(data.Departements):
    d = data[data.Departements == dep]
    departements +=[{"code": dep, 
                     "region": d.Regions.iloc[0],
                     "name" : d.Departements_names.iloc[0],
                     "pop":{ "Pop2013" : {"value" : sum(d["2013"]), "date":2013 },
                             "Pop2008" : {"value" : sum(d["2008"]), "date":2008 },
                             "Pop1999" : {"value" : sum(d["1999"]), "date":1999 },
                             "Pop1990" : {"value" : sum(d["1990"]), "date":1990 },
                             "Pop1982" : {"value" : sum(d["1982"]), "date":1982 },
                             "Pop1975" : {"value" : sum(d["1975"]), "date":1975 },
                             "Pop1968" : {"value" : sum(d["1968"]), "date":1968 },
                             "Pop1962" : {"value" : sum(d["1962"]), "date":1962 } }
                    }]
```

```python
with open('population.json', 'w') as f: #this data file is used in: 
     json.dump({"list": departements}, f)
```
